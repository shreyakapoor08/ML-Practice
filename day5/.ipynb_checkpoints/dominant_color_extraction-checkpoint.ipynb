```python
from sklearn.cluster import KMeans
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./doctor-strange-2-benedict-cumberbatch.jpg") #displays image into bgr format
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(5,5))
plt.imshow(img)
plt.show()
print (img.shape)
```

```python
#flatten the image or reshape image into linear array, we will flatten each channel of the image
c_img = img.copy()
c_img = np.reshape(c_img, (-1,3))

print (c_img.shape)
```

```python
kmeans = KMeans(n_clusters=7,random_state=2) #number of clusters = number of dominant color
```

```python
kmeans.fit_predict(c_img) #fit function will try to map
```

```python
centers = kmeans.cluster_centers_.astype(int) #converting cluster centre values which are in floating point into integer
centers = np.array(centers,dtype='uint8')
print (centers) #each row values represnt rgb value of diff colors
```

# Plot what all colors are these

```python
i = 1

plt.figure(0, figsize=(8,2))

colors = []

for each_col in centers:
    plt.subplot(1,7,i)
    plt.axis("off")
    i+=1
    
    colors.append(each_col)
    
    #color swatch
    a = np.zeros((100,100,3),dtype='uint8')
    a[:,:,:] = each_col
    plt.imshow(a)
    
plt.show()
```

```python
per = np.array(np.unique(kmeans.labels_, return_counts=True)[1], dtype=np.float32)
per = per/c_img.shape[0]

print(per)
```

```python
dom = [ [per[ix], centers[ix]] for ix in range(kmeans.n_clusters) ]
#print (dom)
DOM = sorted(dom, reverse=True)

print (DOM)
```

```python
#creating a color palatte in which the color length will be equal to the percentage
color_p = np.zeros((50,500,3)).astype(int)

start = 0
for ix in range(kmeans.n_clusters):
    width = int( (DOM[ix][0])*color_p.shape[1] )
    
    end = start+width
    color_p[:,start:end, :] = DOM[ix][1]
    start = end
    
plt.imshow(color_p)
plt.show()
```

```python
for ix in range(c_img.shape[0]):
    c_img[ix] = centers[kmeans.labels_[ix]] #for every int predicted we are going to find what is the mapped color
    
c_img = np.reshape(c_img, (img.shape[0], img.shape[1], 3)) #reshape it into the original shape
```

```python

plt.subplot(1,2,1)
plt.axis("off")
plt.imshow(img)

plt.subplot(1,2,2)
plt.axis("off")
plt.imshow(c_img)

plt.show()
```
