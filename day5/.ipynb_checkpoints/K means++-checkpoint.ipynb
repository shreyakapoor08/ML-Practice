K-Means++
1. K-Means is quite sensitive to initialization, if the init is not good, our algo is not able to make desired number of clusters.
2. To overcome this problem, we use the technique called K-Means++ which chooses initial centers so that they are statiscally  close to final ones.
3. The method is default choice for sklearn, so we will use it directly without going into much maths

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(X,y)
```

```python
centers = kmeans.cluster_centers_
```

```python
pred = kmeans.labels_
```

```python
print(pred)
```

```python
plt.scatter(X[:,0],X[:,1],c=pred)
plt.scatter(centers[:,0],centers[:,1],marker='*',color='black')
plt.title("Prediction")
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.title("Actual")
plt.show()
```
