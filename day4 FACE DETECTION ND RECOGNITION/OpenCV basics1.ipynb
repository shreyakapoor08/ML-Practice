```python
#it is a library used when we are working with images, to read and write images, to input a video stream
import cv2
```

```python
#read an image

img = cv2.imread('dog.png') #when we read img by opencv it read image as BGR  instead of RGB and this is bydefault

#switch from BGR to RGB
newImg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
#use matplotlib to show img
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
plt.show()

plt.imshow(newImg)
plt.show()
```
