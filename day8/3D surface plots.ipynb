# 3D Surface Plots

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
#generate a surface
a = [1,2,3] #repeat row wise the number of b times
b = [4,5,6,7] #repeat col wise the number of a times

a,b = np.meshgrid(a,b)

print(a)
print(b)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
plt.show()
```
