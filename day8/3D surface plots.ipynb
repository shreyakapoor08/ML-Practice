# 3D Surface Plots

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
#generate a surface
a = np.arange(-4,4,0.02) #repeat row wise the number of b times
b = a #repeat col wise the number of a times

a,b = np.meshgrid(a,b)

print(a)
print(b)
```

```python
m = a.shape[0]
n = a.shape[1]
print(m,n)

for i in range(m):
    for j in range(n):
        print("(%d,%d)"%(b[i,j],a[i,j]),end=" ")
    print()
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(a,b,a**2+b**2,cmap='rainbow') #countour repre if you hv a 3d bowl and we get planes through the surface, the boundary lines are contour
plt.show()
```
